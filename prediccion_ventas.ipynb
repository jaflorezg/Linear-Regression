# Predicción de Ventas con Regresión Lineal

Supongamos que una tienda quiere predecir sus ventas semanales en función de su inversión en publicidad. Este ejemplo muestra cómo usar regresión lineal para hacer esta predicción.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Datos de ejemplo
data = pd.DataFrame({
    'Publicidad': [1, 2, 3, 4, 5, 6],
    'Ventas': [1.2, 1.9, 3.0, 3.8, 5.1, 6.1]
})
```

```python
# Regresión lineal
X = data[['Publicidad']]  # <- forma segura
y = data['Ventas']

modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Predicción
data['Ventas_Predichas'] = modelo.predict(X)
```

```python
# Visualización
plt.scatter(data['Publicidad'].values, data['Ventas'].values, label='Datos reales') # Apply .values here
plt.plot(data['Publicidad'].values, data['Ventas_Predichas'].values, color='red', label='Línea de regresión') # And here
plt.xlabel('Publicidad (miles de $)')
plt.ylabel('Ventas (miles de $)')
plt.title('Regresión Lineal: Publicidad vs Ventas')
plt.legend()
plt.grid(True)
plt.show()
```
